1. Загрузите файл HR.csv в pandas dataframe

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import shapiro
df = pd.read_csv('HR.csv')
df.head()
```

3. Рассчитайте и визуализируйте корреляционную матрицу для
количественных переменных.
Определите две самые скоррелированные и две наименее
скоррелированные переменные.

```python
df_corr = pd.DataFrame(df, columns=['satisfaction_level',	'last_evaluation', 'average_montly_hours', 'time_spend_company'])
df_corr.corr().round(2)
```

Таким образом, наименее скоррелированные переменные: time_spend_company и satisfaction_level. То есть Стаж работы в компании в годах в меньшей степени зависит от Уровня удовлетворенности работой.

Наиболее скоррелированные переменные: average_montly_hours и last_evaluation. То есть Среднее количество часов на рабочем месте в месяц в большей степени зависит от Времени с момента последней оценки в годах

4. Рассчитайте сколько сотрудников работает в каждом
департаменте.

```python
cnt_workers_department = df.groupby('department')['satisfaction_level'].count().round().sort_values()
cnt_workers_department
```

Вывод: больше всего сотрудников в sales (продажи)

5. Показать распределение сотрудников по зарплатам.

```python
cnt_workers_salary = df.groupby('salary')['satisfaction_level'].count().round().sort_values()
cnt_workers_salary
```

6. Показать распределение сотрудников по зарплатам в каждом департаменте по отдельности

```python
cnt_workers_salary_to_department = df.groupby(['department','salary'])['satisfaction_level'].count().round().sort_values()
cnt_workers_salary_to_department
```

7. Проверить гипотезу, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом

```python
df.groupby('salary')['average_montly_hours'].sum().round().sort_values()
```

Из распределения можно сделать вывод, что гипотеза не верна, сотрудники с высоким окладом проводят на работе меньше времени, чем сотрудники с низким окладом
